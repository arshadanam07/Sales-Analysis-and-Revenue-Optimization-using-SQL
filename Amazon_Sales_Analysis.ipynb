```python
import pandas as pd
import sqlite3

# Read the CSV file into a DataFrame
df = pd.read_csv('amazon.csv')

# Create a SQLite database and establish a connection
conn = sqlite3.connect('amazon.db')

# Write the DataFrame to a SQL table
df.to_sql('amazon_table', conn, if_exists='replace', index=False)

# Execute SQL queries on the table
query = "SELECT * FROM amazon_table"
result = pd.read_sql(query, conn)

# Display the result
print(result)
```

## Cleaning of Data

```python
#Identification of duplicate entries/rows

column_names = "product_id,product_name,category,discounted_price,actual_price,discount_percentage,rating,rating_count,about_product,user_id,user_name,review_id,review_title,review_content,img_link,product_link"
query = "SELECT *, COUNT(*) AS count FROM amazon_table GROUP BY {columns} HAVING count > 1"
duplicate_rows = pd.read_sql(query.format(columns=column_names), conn)
print("Duplicate Rows:")
print(duplicate_rows)
```

```python
#Identify missing values

column_names = ['product_id', 'product_name', 'category', 'discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count', 'about_product', 'user_id','user_name','review_id','review_title','review_content','img_link','product_link']
null_checks = " OR ".join(f"NULLIF({column}, '') IS NULL" for column in column_names)
query = f"SELECT * FROM amazon_table WHERE {null_checks}"
missing_values = pd.read_sql(query, conn)
print("Missing Values in the Table:")
display(missing_values)
```

```python
#connecting the cursor
cursor = conn.cursor()
```

```python
#Changing data type

query = """
UPDATE amazon_table
SET discounted_price = REPLACE(REPLACE(discounted_price, '₹', ''), ',', ''),
    actual_price = REPLACE(REPLACE(actual_price, '₹', ''), ',', ''),
    discount_percentage = REPLACE(discount_percentage, '%', ''),
    rating_count = REPLACE(rating_count, ',', '')
"""
cursor.execute(query)
```

```python
# Commit the changes
conn.commit()
```

## INSIGHTS

```python
#TOTAL REVENUE

query = "SELECT SUM(discounted_price) AS total_revenue FROM amazon_table"
result = pd.read_sql(query, conn)
total_revenue = result['total_revenue'].iloc[0]
print("Total Revenue: ₹", total_revenue)
```

```python
# TOP SELLING PRODUCT OUT OF ALL THE CATEGORIES

query = """
SELECT product_name, category, COUNT(*) AS total_sales
FROM amazon_table
GROUP BY product_name, category
ORDER BY total_sales DESC
LIMIT 10
"""
result = pd.read_sql(query, conn)
top_selling_product = result.iloc[0]
print("Top Selling Product:")
print(top_selling_product)
```

```python
query = """
SELECT user_id, SUM(discounted_price) AS total_revenue
FROM amazon_table
GROUP BY user_id
ORDER BY total_revenue DESC
LIMIT 1;
"""
result = pd.read_sql(query, conn)
print("Highest Revenue Customer:")
print(result)
```

```python
#HIGHEST SELLING CATEGORY
query = """
SELECT category, COUNT(*) AS total_sales
FROM amazon_table
GROUP BY category
ORDER BY total_sales DESC
LIMIT 1
"""
result = pd.read_sql(query, conn)
highest_selling_category = result['category'].values[0]
print("Highest Selling Category:", highest_selling_category)

```

```python
#HIGHEST RATED CATEGORY
query = """
SELECT category, AVG(rating) AS average_rating
FROM amazon_table
GROUP BY category
ORDER BY average_rating DESC
LIMIT 1
"""
result = pd.read_sql(query, conn)
highest_rated_category = result['category'].values[0]
print("Highest Rated Category:", highest_rated_category)
```

```python
#TOP SELLING PRODUCT OUT OF ALL CATEGORIES
query = """
SELECT product_name, COUNT(*) AS total_purchases
FROM amazon_table
GROUP BY product_name
ORDER BY total_purchases DESC
LIMIT 1
"""
result = pd.read_sql(query, conn)
top_selling_product = result['product_name'].values[0]
print("Top Selling Product: ", top_selling_product)
```

```python
#HIGHEST REVENUE CUSTOMER
query = """
SELECT user_name, SUM(discounted_price) AS total_revenue
FROM amazon_table
GROUP BY user_name
ORDER BY total_revenue DESC
LIMIT 1
"""
result = pd.read_sql(query, conn)
highest_revenue_customer = result['user_name'].values[0]
print("Highest Revenue Customer: ", highest_revenue_customer)
```

```python
query = """
SELECT user_name, SUM(discounted_price) AS total_revenue
FROM amazon_table
GROUP BY user_name
ORDER BY total_revenue DESC
LIMIT 1
"""
result = pd.read_sql(query, conn)
highest_revenue_customer = result.iloc[0]['user_name']
total_revenue = result.iloc[0]['total_revenue']
print("Highest Revenue Customer:", highest_revenue_customer)
print("Total Revenue:", total_revenue)
```

```python
#COUNT OF UNIQUE PRODUCT SOLD
query = """
SELECT COUNT(DISTINCT product_id) AS unique_product_count, product_name
FROM amazon_table
GROUP BY product_id, product_name
"""
result = pd.read_sql(query, conn)
print("Count of Unique Products Sold:")
display(result)
```

```python
query = """
SELECT COUNT(DISTINCT product_id) AS total_unique_products_sold
FROM amazon_table
"""
result = pd.read_sql(query, conn)
print("Sum of Unique Products Sold:")
print(result)
```

```python
#HIGHEST RATED PRODUCT OUT OF ALL CATEGORY
query = """
SELECT product_name, category, rating
FROM amazon_table
WHERE rating = (SELECT MAX(rating) FROM amazon_table)
"""
result = pd.read_sql(query, conn)
print("Highest Rated Product Out of All Categories:")
display(result)
```

```python
#HIGHEST RATED PRODUCT IN EACH CATEGORY
query = """
SELECT category, product_name, rating
FROM (
    SELECT category, product_name, rating,
           ROW_NUMBER() OVER (PARTITION BY category ORDER BY rating DESC) AS rank
    FROM amazon_table
) t
WHERE rank = 1
"""
result = pd.read_sql(query, conn)
print("Highest Rated Product in Each Category:")
display(result)
```

```python
#HIGHEST DISCOUNT GIVEN TO WHICH CATEGORY
query = """
SELECT category, MAX(discount_percentage) AS max_discount
FROM amazon_table
GROUP BY category
ORDER BY max_discount DESC
LIMIT 1
"""
result = pd.read_sql(query, conn)
print("Category with the Highest Discount:")
display(result)
```

```python
#HIGHEST DISCOUNT PRODUCT OF ALL CATEGORY
query = """
SELECT product_name, discount_percentage
FROM amazon_table
ORDER BY discount_percentage DESC
LIMIT 1
"""
result = pd.read_sql(query, conn)
print("Product with the Highest Discount (Across All Categories):")
display(result)
```

```python
#HIGHEST DISCOUNT PRODUCT OF EACH CATEGORY

query = """
SELECT category, product_name, discount_percentage
FROM (
    SELECT category, product_name, discount_percentage,
           ROW_NUMBER() OVER (PARTITION BY category ORDER BY discount_percentage DESC) AS rank
    FROM amazon_table
) t
WHERE rank = 1
"""
result = pd.read_sql(query, conn)
print("Product with the Highest Discount in Each Category:")
display(result)
```

```python
#User with the Maximum Number of Products Bought
query = """
SELECT user_id, COUNT(*) AS total_products_bought
FROM amazon_table
GROUP BY user_id
ORDER BY total_products_bought DESC
LIMIT 1
"""
result = pd.read_sql(query, conn)
print("User with the Maximum Number of Products Bought:")
print(result)
```

```python
query = """
SELECT user_id, user_name, COUNT(*) AS total_products_bought
FROM amazon_table
GROUP BY user_id, user_name
HAVING COUNT(*) = (
    SELECT COUNT(*)
    FROM amazon_table
    GROUP BY user_id
    ORDER BY COUNT(*) DESC
    LIMIT 1
)
"""
result = pd.read_sql(query, conn)
print("User with the Maximum Number of Products Bought:")
display(result)
```

## EDA

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Bivariate between category(categorical) and rating(numerical)
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='category', y='rating')
plt.title('Bivariate Bar Graph: Category vs. Rating')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=data['discounted_price'], shade=True)
plt.title('KDE Plot: Discounted Price')
plt.xlabel('Discounted Price')
plt.ylabel('Density')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='discounted_price', kde=True)
plt.title('Histogram: Discounted Price')
plt.xlabel('Discounted Price')
plt.ylabel('Frequency')
plt.tight_layout()

# Show the plot
plt.show()
```
